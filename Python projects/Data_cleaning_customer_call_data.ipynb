```python
#importing the data from a xlsx file i.e excel
import pandas as pd
df = pd.read_excel('Customer Call List.xlsx')
```

```python
#Cleaning it from looking for duplicates and then removing the duplicates

df.count()
#df.info()
df.shape

df = df.drop_duplicates()       #here specifying the df so that the changes gets saved

df
#df.shape                        #checking the changes
```

```python
#dropping columns which is not needed

df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
#striping the data particular to the column which got anomaly related to format

df['Last_Name']

df['Last_Name']=df['Last_Name'].str.strip('123./_') #by this way we are specifying what to remove at once
```

```python
df['Last_Name']
```

```python
#formating the phone numbers
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')

#df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])  # its throwed error as the those are just numbers not string

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))     #making all the numbers strings

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) #formating with -

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
#REplacing the values and standarsing in few columns
#df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df
```

```python
#more replacing overall
df = df.replace('N/a','')
df.fillna('')
```

```python
#dropping the rows which is saying yes to donot contact
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
df = df.fillna('')
df

#similarly removing the rows without phone numbers
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df
```

```python
#lastly ressetting the index
df = df.reset_index(drop = True)
df
```
