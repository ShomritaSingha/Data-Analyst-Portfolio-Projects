```python
#importing the libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing the csv file
df1 = pd.read_csv('world_population.csv')
df1

df1.head(10)

df1[df1['Country'] =='China']    #filtering and sorting with conditions
df1.filter( items=['Country'])
```

```python
#sorting a bit to see the top 2022 population

df1.sort_values(by="2022 Population", ascending= False).head(10)
```

```python
df1.corr()  #for correlation
```

```python
#creating a heatmap to visualise the correlation among the columns
sns.heatmap(df1.corr(), annot = True)

plt.rcParams['figure.figsize'] = (30,7)

plt.show()
```

```python
df1.groupby('Continent').mean().sort_values(by="2022 Population",ascending=False)
```
